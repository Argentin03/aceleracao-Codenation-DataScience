## Git

Git guarda as informações em 3 lugares:

* gitconfig - sistema como um todo
* gitconfig - usuário
* gitconfig - projeto específico de um repositório

Configurações básica:

    git config --global user.name "CarlosChiarelli"

    git config --global usar.email "email_cadastrado_github@hotmail.com"

    git config --global core.editor

Para saber qual é o user configurado no prompt:

    git config --list

### Repositório

Vou até a pasta e 

    git init 

Assim cria uma pasta .git dentro dessa pasta com configurações.

O git possui estados de vida dos arquivos:

* untracked - arquivo foi adicionado ao meu diretório local mas o git ainda não enxerga
* unmodified - arquivo adicionado ao git, não modificado
* modified - se eu modifica-lo, passa para esse estado
* staged - depois de modificar posso jogar ele pra uma área onde será criada a versão

Depois que fazer commit, uma versão do staged é fechada e os arquivos voltam a unmodified.

Diz o status atual do git:

    git status
    
Para passar todos arquivos para unmodified:

    git add *

Salvando a versão junto de uma mensagem para log:

    git commit -m "primeiro Commit de todos arquivos"
    
### Visualizando o histórico

Visualizando histórico de log

    git log
    
Mostra se houve merge, de qual branch foi pra qual branch

    git log --decorate
    
Posso filtra por autor (apenas um pedaço do nome)

    git log --author="Carlos"

Para ver autores em ordem alfabética, quantos commits fizeram e o que dizeram

    git shortlog
    
    # quantidade commits e quem comitou
    git shortlog -sn
    
Mostra em forma gráfica o que está acontecendo com os branchs e as versões

    git log --graph

Dentro do *git log* existe uma hash (número de indentificação), por ela consigo saber o que aconteceu no commit

    git show numeroHash
    
Sempre preciso usar o *git diff* antes de fazer um commit para ver as diferenças, dou uma conferida antes de salvar 

    git diff
    
    # somente arquivos modificados 
    git diff --name-only

### Desfazendo coisas

Possibilidade de resetar os erros, tirar coisas específicas.

    # retorna arquivo para antes da edição (desfaz tudo feito)
    git checkout arquivo

Quando quero tirar um arquivo do estado de stage que estou:

    git reset HEAD arquivo

### Repositório criado manualmente (cliques no Github)

Após o repositório ser criado no Github, deve-se mapea-lo na máquina com a SSH.

    git remote add origin git@github.com:CarlosChiarelli/desafiosCodenationDataScience.git
    
Para ver informações sobre o repositório remoto

    git remote -v
    
Enviar os arquivos que eu tenho e todas informações (logs, histórico) para repositório remoto
    
    # foi esse comando que funcionou (inicialmente)
    git push -f origin master
    
    # caso o de cima não rode, executar esse antes
    git pull origin master
    
    # foi sugerido executar apenas esse mas não funciona
    git push -u origin master 
    
Feito isso pela primeira vez, para subir arquivos EDITADOS (já devem ter sido adicionados) basta 

    # comittar a mudança no remoto
    git commit -am "descricao" 
    
    git push origin master 
    
### Clonando repositório remoto

Para copiar um Github para pasta local na máquina se deve pegar o endereço dele (SSH) e 

    git clone endereco_ssh nome_pasta_nova

### Fork

Pegar um projeto que não é meu e faz uma cópia dele pra mim. Primeiro copio o repositório pra mim, faço as mudanças e depois mando pra ela. O fork (feito manualmente) permite eu clonar diretamente para meu github, mas para isso preciso de organizações criadas, onde vou salvar os arvuivos do github alvo.
